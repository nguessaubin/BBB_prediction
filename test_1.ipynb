```python
def Search_duplicates(mol_list):
    " Give mol list to detect doublons"
    dupl_list=[]
    j=0
    length = len(mol_list)
    for i in range(length-1):
        for j in range(i+1,length):
            if mol_list[i]==mol_list[j]:
                dupl_list.append(j) 
    return dupl_list
```
