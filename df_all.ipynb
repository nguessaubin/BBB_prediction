```python
df1 = pd.read_csv('df_after_PARSE1.csv')
df1.set_index('index',inplace=True)
```

```python
from rdkit import DataStructs
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
mol_list = df1.index
j=0
i=0
tanimoto=()
Tan_mol_list=[]
length = len(df1.index)
for i in range(length):
    for j in range(length):
        p =DataStructs.TanimotoSimilarity(Chem.RDKFingerprint(Chem.MolFromSmiles(mol_list[i])),Chem.RDKFingerprint(Chem.MolFromSmiles(mol_list[j])))
        if (p==1.0) & (i!=j):
            Tan_mol_list.append(mol_list[j])
        p=0
        k=0
    for k in range(len(Tan_mol_list)) :
        if df.loc[mol_list[i]]['logBB']!=df.loc[Tan_mol_list[k]]['logBB']:
            tanimoto +=(mol_list[i],Tan_mol_list)
            break
    Tan_mol_list=[]
```

```python
mol_list_1 = list(df1.index)
Tan_mol_list=['O=C1NC(=O)C(c2ccccc2)(c2ccccc2)N1','O=C1NC(C(C2=CC=CC=C2)(C3=CC=CC=C3)N1)=O']
for mol in Tan_mol_list:
    mol_list_1.remove(mol)
    
len(mol_list_1)
```

```python
similarity = True
mol_list_1 = list(df1.index)
Tan_mol_list=[]
tanimoto1=()
j=0
while similarity:
    for i in range(len(mol_list_1)-1):
        p =DataStructs.TanimotoSimilarity(Chem.RDKFingerprint(Chem.MolFromSmiles(mol_list_1[0])),Chem.RDKFingerprint(Chem.MolFromSmiles(mol_list_1[i+1])))
        if p==1.0:
            Tan_mol_list.append(mol_list_1[i+1])
        
    p=0
    k=0
    for k in range(len(Tan_mol_list)) :
        if df.loc[mol_list_1[0]]['logBB']!=df.loc[Tan_mol_list[k]]['logBB']:
            tanimoto1 +=(mol_list_1[0],Tan_mol_list)
            break
    for mol in Tan_mol_list:
        mol_list_1.remove(mol)
    Tan_mol_list = []
    mol_list_1.remove(mol_list_1[0])
    if len(mol_list_1)==0:
        similarity =False
    
```
